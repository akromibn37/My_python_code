```python
import requests
from  bs4 import BeautifulSoup
```

```python
r=requests.get("http://www.soccersuck.com")
c=r.content
```

```python
soup=BeautifulSoup(c,"html.parser")
print(soup.prettify)
```

```python
all=soup.find_all("div",{"class":"headline_big_video"})
all
```

```python
for i in all:
    print(i.find_all("img")[0].text)
```

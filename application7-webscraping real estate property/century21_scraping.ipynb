```python
import requests
from  bs4 import BeautifulSoup
r2=requests.get("http://www.century21.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
c2=r2.content
soup2=BeautifulSoup(c2,"html.parser")

all2=soup2.find_all("div",{"class":"property-card-primary-info"})
```

```python
all2[0].find_all("a",{"class":"listing-price"})[0].text.replace(" ","").replace("\n","")
```

```python
for i in all2:        
        print(i.find("a",{"class":"listing-price"}).text.replace(" ","").replace("\n",""))
        print(i.find_all("div",{"class":"property-address"})[0].text.replace(" ",""))
        print(i.find_all("div",{"class":"property-city"})[0].text.replace(" ",""))
        try:
            print(i.find("div",{"class":"property-beds"}).text)
        except:
            pass
        print("------------")
```

```python
l=[]
for i in all2:
        d={}
        d["Price"]=i.find("a",{"class":"listing-price"}).text.replace(" ","").replace("\n","")
        d["Address"]=i.find_all("div",{"class":"property-address"})[0].text.replace(" ","").replace("\n","")
        d["Locality"]=i.find_all("div",{"class":"property-city"})[0].text.replace(" ","").replace("\n","")

        try:
            d["Beds"]=i.find_all("div",{"class":"property-beds"})[0].text.replace("beds","").replace(" ","").replace("\n","")
        except:
             d["Beds"]=None
#         try:
#              d["Baths"]=i.find_all("div",{"class":"property-baths"})[0].text.replace("bath","").replace("\n","")
#         except:
#              d["Baths"]=None
        try:
             d["Area"]=i.find_all("div",{"class":"property-sqft"})[0].text.replace("sq. ft","").replace("\n","").replace(" ","")
        except:
             d["Area"]=None
        l.append(d)
```

```python
import pandas
df=pandas.DataFrame(l)
df
```

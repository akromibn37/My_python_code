```python
import requests
from bs4 import BeautifulSoup

r=requests.get("http://www.priceza.com/search?category=0&brand=Apple&page=1&sort=1&productdataname=iphone%206s")
c=r.content
soup=BeautifulSoup(c,"html.parser")
print(soup.prettify())
```

```python
all=soup.find_all("div",{"class","grid-box"})
len(all)
```

```python
all2=soup.find_all("div",{"class","pd-info"})
print(all2[0].find("span",{"itemprop":"description"}).text.replace("\t","").replace("\n","").replace("\r",""))
all3=soup.find_all("div",{"class":"bullet-spec"})
print(all3[0].find("ul").text)
len(all2)
len(all3)
```

```python
l=[]
for i,j in zip(all,all2):
    d={}
    d["Product name"]=i.find_all("h3",{"itemprop":"name"})[0].text
    try:
        d["Price"]=i.find_all("span",{"itemprop":"price"})[0].text.replace("฿","").replace("\n","")
    except:
        d["Price"]=None
    d["Description"]=j.find("span",{"itemprop":"description"}).text.replace("\t","").replace("\n","").replace("\r","")
#     d["Battery"]=k.find_all("li")[0].text.replace("Battery: ","")
#     d["Camera"]=k.find_all("li")[1].text.replace("Camera: ","")
#     d["Screen"]=k.find_all("li")[2].text.replace("Screen Size: ","")
#     d["Storage"]=k.find_all("li")[3].text.replace("Storage: ","")   
    l.append(d)
len(l)
```
